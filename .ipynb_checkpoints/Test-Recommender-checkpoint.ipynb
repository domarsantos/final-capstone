```python
import sklearn_recommender as skr
import pandas as pd
```

```python
df = pd.read_csv(r'dataSet.csv')
```

```python
df.head(10)
```

```python
df.drop(columns=['Population','AreaPerSqMeter','HabitantPerHA','PopGrowthRatePerYear','Latitude','Longitude'],inplace=True)
```

```python
df.head(10)
```

```python
df.shape
```

```python
# df[df.columns[1]]
```

```python
tf = skr.transformer.SimilarityTransformer(cols=(1, -1), normalize=True)
sim_mat = tf.transform(df)
```

```python
sim_mat.shape
```

```python
rec = skr.recommender.SimilarityRecommender(5)
rec.fit(sim_mat)
# predict the 5 most similar items to the given items 5, 6 and 7 respectively
rec.predict([5, 8])
```
